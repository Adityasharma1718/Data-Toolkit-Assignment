1. What is NumPy, and why is it widely used in Python?

- NumPy (Numerical Python) is a library for numerical and scientific computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently. It is widely used for its speed, memory efficiency, and integration with other libraries.

2. How does broadcasting work in NumPy?


- Broadcasting allows NumPy to perform element-wise operations on arrays of different shapes by automatically expanding the smaller array to match the larger one. It simplifies coding and improves performance without requiring explicit replication of data.

3. What is a Pandas DataFrame?


-
A DataFrame is a two-dimensional, size-mutable, and heterogeneous tabular data structure with labeled axes (rows and columns). It is a core data structure in the Pandas library, similar to a spreadsheet or SQL table.

4. Explain the use of the groupby() method in Pandas.


-
The groupby() method splits data into groups based on some criteria (e.g., column values), applies a function to each group (e.g., mean, sum), and combines the results. It is useful for aggregation and transformation operations.

5. Why is Seaborn preferred for statistical visualizations?


-
Seaborn is built on Matplotlib and provides high-level, aesthetically pleasing functions for complex statistical visualizations such as box plots, violin plots, heatmaps, and regression plots with minimal code.

6. What are the differences between NumPy arrays and Python lists?


-
Speed: NumPy arrays are faster.

Memory: Arrays use less memory.

Functionality: Arrays support vectorized operations; lists do not.

Data type: Arrays are homogeneous; lists can be heterogeneous.

7. What is a heatmap, and when should it be used?


-
A heatmap is a graphical representation of data where values are represented with color. It is used to visualize correlation matrices, data density, or missing values across rows and columns.

8. What does the term “vectorized operation” mean in NumPy?


-
Vectorized operations allow you to apply operations to entire arrays without writing explicit loops. This leads to cleaner code and faster execution due to optimized C-based back-end.

9. How does Matplotlib differ from Plotly?


-
Matplotlib is static and suitable for simple 2D plots.

Plotly is interactive and supports dynamic, zoomable, and web-based plots.
Plotly is often preferred for dashboards and web applications.

10. What is the significance of hierarchical indexing in Pandas?


-
Hierarchical indexing (MultiIndex) allows you to have multiple (two or more) index levels on a DataFrame or Series, enabling complex data manipulation and reshaping.

11. What is the role of Seaborn’s pairplot() function?


-
pairplot() creates a matrix of scatter plots for each pair of numerical columns in a DataFrame and optionally includes histograms or KDE plots on the diagonals. It helps explore relationships between variables.

12. What is the purpose of the describe() function in Pandas?


-
The describe() function generates descriptive statistics (like count, mean, std, min, max, quartiles) for numeric columns, helping in quick data analysis.

13. Why is handling missing data important in Pandas?


-
Missing data can lead to incorrect or misleading results in analysis. Pandas provides tools (dropna(), fillna()) to detect, remove, or impute missing values, ensuring data quality.

14. What are the benefits of using Plotly for data visualization?


-
Interactive and dynamic plots

Web-based rendering

Supports complex visualizations (e.g., 3D, maps)

Easy integration with dashboards and notebooks

15. How does NumPy handle multidimensional arrays?


-
NumPy uses the ndarray object to handle multidimensional arrays, supporting indexing, slicing, reshaping, and broadcasting across any number of dimensions (e.g., 1D, 2D, 3D, etc.).

16. What is the role of Bokeh in data visualization?


-
Bokeh is a Python interactive visualization library that renders HTML and JavaScript plots. It supports real-time updates, interactivity, and is often used in dashboards and web apps.

17. Explain the difference between apply() and map() in Pandas.


-
map() is used for element-wise transformations on Series.

apply() can be used on both Series and DataFrames, allowing application of functions row-wise or column-wise.

18. What are some advanced features of NumPy?


-
Broadcasting and vectorized operations

Linear algebra and Fourier transforms

Masking and conditional logic

Memory-mapped files for large datasets

Structured arrays and recarrays

19. How does Pandas simplify time series analysis?


-
Pandas offers:

Date and time indexing (DatetimeIndex)

Resampling, shifting, rolling statistics

Built-in handling of datetime formats

Easy frequency conversion

20. What is the role of a pivot table in Pandas?


-
A pivot table summarizes and reshapes data using aggregation. It allows users to group data by one or more keys and apply functions like sum, mean, or count to get a cross-tabulated view.

21. Why is NumPy’s array slicing faster than Python’s list slicing?


-
NumPy arrays are stored in contiguous memory blocks and use low-level optimizations. Python lists are more flexible but less efficient, as they store references to objects.

22. What are some common use cases for Seaborn?


-
Exploratory data analysis (EDA)

Correlation heatmaps

Distribution plots (e.g., histograms, KDE)

Categorical plots (box, violin, bar)

Pairwise relationships visualization



1. How do you create a 2D NumPy array and calculate the sum of each row?

python

```python
#3. Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()


```

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'Math': [90, 85, 88, 92],
    'Science': [85, 80, 90, 95],
    'English': [78, 75, 80, 82]
})
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
#5. Generate a bar plot using Plotly.
import plotly.express as px
import pandas as pd

df = pd.DataFrame({'Fruits': ['Apple', 'Banana', 'Mango'], 'Count': [10, 15, 7]})
fig = px.bar(df, x='Fruits', y='Count', title='Fruit Count')
fig.show()
```

```python
#6. Create a DataFrame and add a new column based on an existing column.
import pandas as pd

df = pd.DataFrame({'Price': [100, 200, 300]})
df['Price_With_Tax'] = df['Price'] * 1.18
print(df)
```

```python
#7- Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print(result)
```

```python
#8- Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [5, 15, 25, 35]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title('Multiple Line Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#9- Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

df = pd.DataFrame({'Score': [45, 78, 62, 90]})
filtered_df = df[df['Score'] > 70]
print(filtered_df)
```

```python
#10- Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 30, 40, 40, 50]
sns.histplot(data, kde=True)
plt.show()
```

```python
#11. Perform matrix multiplication using NumPy.
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```

```python
#13- Create a 3D scatter plot using Plotly
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 15, 20, 25],
    'z': [100, 200, 300, 400]
})

fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
